```python
# Create a list of synthetic data
data = [10, 20, 30, 40, 50, 60, 70]

# Calculate the mean of the data
mean = sum(data)/len(data)
print('Mean =', mean)

# Standardize the data
standardized_data = [(x - mean)/(max(data) - min(data)) for x in data]

# Print out the standardized data
print('Standardized Data =', standardized_data)
```
